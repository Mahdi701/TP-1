```python
import pandas as pd
df=pd.read_csv('titanic-passengers.csv' ,sep=';')
df["Cabin"].fillna('G6' ,inplace=True)
df.drop(['Cabin', 'Name'], axis=1 ,inplace=True)
df['Age'].fillna(df['Age'].median() ,inplace=True)
df.tail()
```
